### Task-1 Data Parsing & Profile Cleaning

```python
# Raw Data

raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

```python
students_data = []

for student in raw_students :
    name1 = student["name"].strip().title() #trimming invalid spaces and converting into Title Case.
    roll1 = int(student["roll"]) #converting the data type of roll number from string to interger.
    marks_list = student["marks_str"].split(", ") #spliting the marks on the basis of "," and storing them in separte variable named as marks_split.
    marks1 = [] #converting the data type of splitted marks from string to integer.
    for mark in marks_list:
        marks1.append(int(mark))

    cleaned_data = { #saving cleaned data to students_data
        "name"  : name1,
        "roll"  : roll1,
        "marks" : marks1
    }

    students_data.append(cleaned_data)

    words = name1.split()
    is_valid = True
    for word in words:
        if not word.isalpha():
            is_valid = False
            break

    if not word.isalpha():
            is_valid = False
            break

    if is_valid:
        validity = "Valid name"
    else:
        validity = "Invalid name"

    print("================================")
    print(f"Student : {name1}   {validity}")
    print(f"Roll No : {roll1}")
    print(f"Marks   : {marks1}")
    print("================================")
    print()
```

```python
for student in students_data:
    if student["roll"] == 103:
        print(f"Roll 103 Name in ALL CAPS  : {student['name'].upper()}")
        print(f"Roll 103 Name in lowercase : {student['name'].lower()}")
```

### Task 2 - Marks Analysis Using Loops & Conditionals

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
```

```python
print(f"Grade Report for: {student_name}")
print("------------------------------------")

for i in range(len(subjects)):
    mark = marks[i]
    subject = subjects[i]

    # Check which grade range the mark falls in
    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"

    print(f"{subject} : {mark} --> Grade: {grade}")
```

```python
total = 0
for mark in marks:
    total = total + mark

average = round(total / len(marks), 2)

print(f"\nTotal Marks   : {total}")
print(f"Average Marks : {average}")
```

```python
highest_mark = marks[0]
lowest_mark  = marks[0]
highest_sub  = subjects[0]
lowest_sub   = subjects[0]

for i in range(len(marks)):
    if marks[i] > highest_mark:
        highest_mark = marks[i]
        highest_sub  = subjects[i]

    if marks[i] < lowest_mark:
        lowest_mark = marks[i]
        lowest_sub  = subjects[i]

print(f"Highest Subject : {highest_sub} ({highest_mark})")
print(f"Lowest Subject  : {lowest_sub} ({lowest_mark})")
```

```python
print("\n--- Add New Subjects ---")
print("Type 'done' to stop adding subjects")

new_subjects_count = 0

while True:

    sub_name = input("Enter subject name (or 'done' to stop): ")

    if sub_name.lower() == "done":
        break

    mark_input = input(f"Enter marks for {sub_name} (0-100): ")

    if not mark_input.isdigit():
        print("Warning: Invalid input! Marks must be a number. Skipping.")
        continue

    new_mark = int(mark_input)

    if new_mark < 0 or new_mark > 100:
        print("Warning: Marks must be between 0 and 100. Skipping.")
        continue

    subjects.append(sub_name)
    marks.append(new_mark)
    new_subjects_count = new_subjects_count + 1
    print(f"{sub_name} with marks {new_mark} added successfully!")
```

```python
updated_total = 0
for mark in marks:
    updated_total = updated_total + mark

updated_average = round(updated_total / len(marks), 2)

print(f"\nNew subjects added     : {new_subjects_count}")
print(f"Updated Average Marks  : {updated_average}")
```
